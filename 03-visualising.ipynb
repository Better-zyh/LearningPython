# Subplot

```python
import numpy as np
import matplotlib.pyplot as plt
arthritis=np.loadtxt(fname='data/inflammation-01.csv',delimiter=',')
figure=plt.figure(figsize=(10,3))
axes1=figure.add_subplot(1,3,1) #1 row, 3 column, 1st graph
axes2=figure.add_subplot(1,3,2) 
axes3=figure.add_subplot(1,3,3) 
axes1.set_ylabel('average')
axes1.plot(arthritis.mean(axis=0))
axes2.set_ylabel('max')
axes2.plot(arthritis.max(axis=0))
axes3.set_ylabel('min')
axes3.plot(arthritis.min(axis=0))
figure.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
arthritis=np.loadtxt(fname='data/inflammation-01.csv',delimiter=',')
figure=plt.figure(figsize=(3,10))
axes1=figure.add_subplot(3,1,1) 
axes2=figure.add_subplot(3,1,2) 
axes3=figure.add_subplot(3,1,3) 
axes1.set_ylabel('average')
axes1.plot(arthritis.mean(axis=0))
axes2.set_ylabel('max')
axes2.plot(arthritis.max(axis=0))
axes3.set_ylabel('min')
axes3.plot(arthritis.min(axis=0))
figure.tight_layout()
plt.show()
```
